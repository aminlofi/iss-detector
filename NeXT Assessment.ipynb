```python
# compile all into one function
def get_iss_info(tm):
    
    df_storedate = pd.DataFrame(columns=['Date'])
    
    df_store1 = pd.DataFrame()
    df_store2 = pd.DataFrame()
    
    ts_array = []
    ts_arrayb = []
    
    # store timeframe in datetime for user readability
    dateformat = []
    dateformat_b = []
    
    element = datetime.datetime.strptime(tm, "%d-%m-%Y %H:%M:%S")

    timestamp = datetime.datetime.timestamp(element)
    
    timestampb = timestamp
    
    ts_array.append(timestamp)
    ts_arrayb.append(timestampb)
    
    # 60 for cycle 60 minutes (1 hour)
    for i in range(2):
        # 60 seconds (1 minute)
        # move back 60 minutes
        timestamp = timestamp + 60
        ts_array.append(timestamp)
        print(ts_array)
        # move forward 60 minutes
        timestampb = timestampb - 60
        ts_arrayb.append(timestampb)
        print(ts_arrayb)
    
    for i in ts_array:
        new_date = datetime.datetime.utcfromtimestamp(i).strftime('%d-%m-%Y %H:%M:%S')
        dateformat.append(new_date)
    
    df_storedate['Date'] = dateformat
    
    for i in ts_arrayb:
        new_dateb = datetime.datetime.utcfromtimestamp(i).strftime('%d-%m-%Y %H:%M:%S')
        dateformat_b.append(new_dateb)
    
    df_storedate['Date'] = dateformat_b
     
    for x in ts_array:
        query = {'id':'25544', 'timestamp':x}
        response = requests.get('https://api.wheretheiss.at/v1/satellites', params=query).json()
        df = pd.DataFrame.from_dict([response])
        df_store1 = df_store1.append(df)
        print(df_store1)
        time.sleep(2)
        
    for y in ts_arrayb:
        query = {'id':'25544', 'timestamp':y}
        response = requests.get('https://api.wheretheiss.at/v1/satellites', params=query).json()
        df2 = pd.DataFrame.from_dict([response])
        df_store2 = df_store2.append(df2)
        print(df_store2)
        time.sleep(2)
    
    df_storefinal = pd.concat([df_store1, df_store2], ignore_index=True)
    df_storefinal['latitude'] = df_storefinal['latitude'].astype(str)
    df_storefinal['longitude'] = df_storefinal['longitude'].astype(str)
        
    for j, rows in df_storefinal.iterrows():
        lat = rows['latitude']
        long = rows['longitude']
        response_b = requests.get('https://api.wheretheiss.at/v1/coordinates/' + lat +  ',' + long).json()
        df3 = pd.DataFrame.from_dict([response_b])
        print(df3)
        df_storefinal = df_storefinal.append(df3)
        time.sleep(2)
    
    df_storefinal = df_storefinal.append(df_storedate)
    
    return (list(df_storefinal['Date']), list(df_storefinal['timezone_id']), list(df_storefinal['latitude']), list(df_storefinal['longitude']))
```

```python
df_new = pd.DataFrame()

app = JupyterDash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP], prevent_initial_callbacks=False)

app.layout = html.Div()

SIDEBAR_STYLE = {
    "position": "fixed",
    "top": 0,
    "left": 0,
    "bottom": 0,
    "width": "26rem",
    "padding": "2rem 1rem",
    "background-color": "#e9ecf5",
    "color": '#FFFFFF',
    "overflow": 'scroll'
}

CONTENT_STYLE = {
    "margin-left": "26rem",
    "marigin-right": "0rem",
    "padding": "2rem 1rem",
    "background-color": "#303540",
    "color": '#FFFFFF',
    "font-color": '#FFFFF'
}

sidebar = html.Div([
    
], style=SIDEBAR_STYLE)

content = html.Div([
    
    html.Div([
        html.Label('Enter Date: '),
        dcc.Input(
            id = 'start-time-input',
            type = 'text',
            value = '12-11-2021 12:19:00'
        ),
    ]),
    
    html.Br(),
    
    dbc.Alert("", color="info", id='confirmed'),
    
    html.Div(children=[
        html.Div(children=[html.H2('Please wait while data is being processed')], 
                 style={'width': '100%', 'display': 'inline-block', 'border': '2px solid black', 'text-align': 'center', 'font-family': 'Century Gothic'})
    ])

], style=CONTENT_STYLE)

app.layout = html.Div([sidebar, content])

@app.callback(
    dash.dependencies.Output('confirmed', 'children'),
    dash.dependencies.Input('start-time-input', 'value'),
)
def update_output(masa):
    a = get_iss_info(masa)
    html.Li(style={"padding": "2rem 1rem"})
    return html.Ul([html.Li(x) for x in a])
    
    
app.run_server(mode='external')
```
